# LOGISTIC REGRESSION

### Logistic Regression is learning technique borrowed by machine learning from statistic.It predicts the probability of occurance.In spite of its name it is not used for regression problem it predicts the binary outcomes.The idea of logistic regression is to find the relationship between features and probability of outcome.Logistic Regression is used when the dependent variable(target) is categorical.

### For example,

    1-To predict whether an email is spam (1) or (0)
   
    2-Whether the tumor is malignant (1) or not (0)



## Types of Logistic Regression

### 1. Binary Logistic Regression
The categorical response has only two 2 possible outcomes. Example: Spam or Not

### 2.Multinomial Logistic Regression
Three or more categories without ordering. Example: Predicting which food is preferred more (Veg, Non-Veg, Vegan)

### 3. Ordinal Logistic Regression
Three or more categories with ordering. Example: Movie rating from 1 to 5

## >>Importing different libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn import preprocessing
```

## >>Read csv file

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
```

## >> Partitioning of input feature data and target data

```python
x = dataset.iloc[:,2:4].values
y = dataset.iloc[:,4].values
```

## >> Split the training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

## >> This is the step for applying Logistic Regression 

```python
model=LogisticRegression()
#simply fit the values of x tarin and y tarin
model.fit(X_train, y_train)
model.score(X_train,y_train)
pred=model.predict(X_test)
print(pred)
```

## >>Accuracy can be measured like this

```python
print("Accuracy",metrics.accuracy_score(y_test,pred))
```
